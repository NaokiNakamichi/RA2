```python
import os
import numpy as np
from pdf2image import convert_from_path
from PIL import Image
import matplotlib.pyplot as plt
```

```python
images = convert_from_path('sample.pdf')
```

```python
sample = images[0].convert("L")

sample
```

```python
k = np.array(sample)
```

```python
k
```

```python
thresh = 128

im_bool = k > thresh
```

```python
maxval = 255

im_bin = (k > thresh) * maxval
```

```python
Image.fromarray((im_bin).astype(np.uint8))
```

```python
for i in range(0,np.shape(im_bin)[0],3):
    tmp = np.mean(im_bin[i+1,:]) - np.mean(im_bin[i,:])
    if tmp < -50:
        print(tmp)
        print(i)

kaklu = np.mean(im_bin,axis=1)
```

```python
plt.hist(kaklu)
plt.show()
```
